# I. GraphScene

Les GraphScene étaient les scènes utilisées dans les précédentes versions de Manim (0.5.0 et inférieures) pour afficher des graphiques et des courbes.

Les **GraphScene deviendront obsolètes** dans les futures mises à jour, et c'est pour cela que je ne vous les présente pas.


# II. Axes

### <center><u>II.1. Un premier graphique</u></center>

Pour ce graphique, on va afficher la simple fonction **f(x) = x**

```python
%%manim FirstAxes

from manim import *

class FirstAxes(Scene):
    def construct(self):
        #On créé l'axe de notre graphique
        axes = Axes(
            x_range = [-10, 10, 1],
            y_range = [-10, 10, 1],
            axis_config = {"include_numbers" : True}
        )
        
        axes.center()
        graph = axes.get_graph(
            lambda x : x,
            color = GOLD_A
        )
        
        self.play(Create(axes))
        self.play(Create(graph))
        self.wait(1)
```

### <center><u>II.2. f(x) = x²</u></center>

```python
%%manim x_squared_graph

from manim import *

class x_squared_graph(Scene):
    def construct(self):
        #On créé l'axe plan de notre graphique
        axes = Axes(
            x_range = [-10, 10, 1],
            y_range = [0, 110, 10],
            axis_config = {"include_numbers" : True}
        )
        
        axes.center()
        graph = axes.get_graph(lambda x : self.x_squared(x), 
                               color=YELLOW)
        
        self.play(Create(axes))
        self.play(Create(graph))
        self.wait(1)
        
    def x_squared(self, x):
        return x**2
```

### <center><u>II.3. f(x) = sin(x)</u></center>

```python
%%manim -s sinus_graph

import math
from manim import *
class sinus_graph(Scene):
    def construct(self):
        #On créé l'axe plan de notre graphique
        axes = Axes(
            x_range = [0, 4*math.pi, math.pi/4],
            y_range = [-1.0, 1.1, 0.1],
            axis_config = {"include_numbers" : True},
            decimal_number_config = {"num_decimal_places" : 3}
        )
        
        axes.center()
        graph = axes.get_graph(lambda x : math.sin(x), 
                               color=GOLD_B)
        
        self.play(Create(axes))
        self.play(Create(graph))
        self.wait(1)
```

# III. Mise en forme des axes graphique

Les **paramètres de mises en forme des axes du graphique** sont ceux utilisés pour les objets **NumberLine**.

Documentation NumberLine : https://docs.manim.community/en/latest/reference/manim.mobject.number_line.NumberLine.html#manim.mobject.number_line.NumberLine

### <center><u>III.1. Modifier la taille des ticks</u></center>

```python
%%manim -s TickGraph

from manim import *
class TickGraph(Scene):
    def construct(self):
        #On créé les axes de notre graphique
        axes = Axes(
            x_range = [0, 11, 1],
            y_range = [0,70000, 5000],
            axis_config = {"include_numbers" : True},
            x_axis_config = {"tick_size" : 0.2},
            y_axis_config = {"tick_size" : 3.5}
        )
        
        #On récupère la fonction exponentielle
        graph = axes.get_graph(lambda x : math.exp(x), color=ORANGE)
        
        #Animation du graphique
        self.play(Create(axes))
        self.play(Create(graph))
        self.wait(1)
```

### <center><u>III.4. Faire pivoter les axes</u></center>

Le paramètre **rotation** prend une valeur en **radians**.

La méthode radians() du module math permet de convertir un angle exprimé en degrés en radians.

```python
%%manim -s RotationGraph

import math
from manim import *
class RotationGraph(Scene):
    def construct(self):
        #On créé les axes de notre graphique
        axes = Axes(
            x_range = [-10, 11, 1],
            y_range = [-10, 11, 1],
            axis_config = {"include_numbers" : True},
            x_axis_config = {"tick_size" : 0.05, "color" : RED, "rotation" : math.radians(-45)},
            y_axis_config = {"tick_size" : 0.05, "color" : YELLOW, "rotation" : math.radians(45)}
        )
        
        #On récupère la fonction exponentielle
        graph = axes.get_graph(lambda x : x, color=BLUE_A)
        
        #Animation du graphique
        self.play(Create(axes))
        self.play(Create(graph))
        self.wait(1)
```

On constate que **la fonction pivote avec les axes**.
